```python
df=pd.read_csv('orders.csv', sep=',', encoding='utf-8', on_bad_lines='warn')
df.head(10)
```

```python
# ad1.
df['order_date']=pd.to_datetime(df["order_date"], format="%Y/%m/%d")
print('data: ', df['order_date'].tail(1))
#ad2
df['tshirt_category'].value_counts()
display(df['tshirt_category'].value_counts())
#ad3. obsługa brakujących danych i ujednolicenie nazw
lista_zmian={"Wh ":"White ", "Bl ":"Black ", "Tshirt":"T-Shirt", "Tennis Shirt":"Tennis Shirt all-size" ,"Hoodie":"no-info Hoodie all-size"}
for key in lista_zmian.keys():
    df['tshirt_category']=df['tshirt_category'].apply(lambda x : x.replace(key, str(lista_zmian[key])))
display('after:',df['tshirt_category'].value_counts())
```

```python
df[[ 'tshirt_color','tshirt_type', 'tshirt_size']]=df['tshirt_category'].apply(lambda x : pd.Series(str(x).split()))
# df[[ 'tshirt_color','tshirt_type', 'tshirt_size']]=df['tshirt_category'].str.split(expand=True) #2 sposob szybszy ale bez lambda czas: 1' (poprzedni 11')
```
